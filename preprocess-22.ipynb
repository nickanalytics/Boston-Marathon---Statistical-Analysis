## Concatenate all files into one dataframe

```python
import pandas as pd

# List of CSV file names
csv_files = ['m-a.csv', 'm-b.csv', 'm-c.csv', 'm-d.csv', 'm-e.csv', 'm-f.csv', 'm-g.csv', 'm-h.csv', 'm-i.csv', 'm-j.csv', 'm-k.csv', 'm-l.csv', 'm-m.csv', 'm-n.csv', 'm-o.csv', 'm-p.csv', 'm-q.csv', 'm-r.csv', 'm-s.csv', 'm-t.csv', 'm-u.csv', 'm-v.csv', 'm-w.csv', 'm-x.csv', 'm-y.csv', 'm-z.csv', 'm-80+.csv', 'm-75+.csv', 'm-70+.csv', 'm-65+.csv', 'm-60+.csv','v-z.csv','v-y.csv','v-x.csv','v-w.csv','v-v.csv','v-u.csv','v-t.csv','v-s.csv','v-r.csv','v-q.csv','v-p.csv','v-o.csv','v-n.csv','v-m.csv','v-l.csv','v-k.csv','v-j.csv','v-i.csv','v-h.csv','v-g.csv','v-f.csv','v-e.csv','v-d.csv','v-c.csv','v-b.csv','v-a.csv','v-75+.csv','v-70+.csv','v-65+.csv','v-60+.csv','v-55+.csv','m-1.csv','m-2.csv','m-3.csv','m-4.csv','m-5.csv','m-6.csv','m-7.csv','m-9.csv','s-1.csv','s-2.csv','s-3.csv','s-4.csv','s-5.csv','s-6.csv','s-7.csv','s-8.csv','s-9.csv','b-1.csv','b-2.csv','b-3.csv','b-4.csv','b-5.csv','b-6.csv','b-7.csv','b-8.csv','b-9.csv','c-1.csv','c-2.csv','c-3.csv','c-4.csv','c-5.csv','c-6.csv','c-7.csv','c-8.csv']

# Loop through each CSV file and read into a DataFrame with dynamically created variable names
for i, file in enumerate(csv_files):
    globals()[f'df{i+1}'] = pd.read_csv(file)
    
    # create a new column called 'key'
    globals()[f'df{i+1}']['key'] = (globals()[f'df{i+1}'].index // 2) + 1

    # create a new column called 'key_order'
    globals()[f'df{i+1}']['key_order'] = ((globals()[f'df{i+1}']['key'] != globals()[f'df{i+1}']['key'].shift()) | (globals()[f'df{i+1}'].index % 2 == 0)).astype(int)

    # select rows where 'key_order' equals 1 and create a new dataframe
    new_df_1 = globals()[f'df{i+1}'].loc[globals()[f'df{i+1}']['key_order'] == 1].reset_index(drop=True)

    # select rows where 'key_order' equals 0 and keep them in the original dataframe
    new_df_0 = globals()[f'df{i+1}'].loc[globals()[f'df{i+1}']['key_order'] == 0].reset_index(drop=True)

    # merge the two dataframes based on the 'key' column
    globals()[f'df{i+1}'] = new_df_1.merge(new_df_0, on='key', how='outer')
```

```python
# create an empty list to store the dataframes
dfs = []

# loop through all the dataframe names and append them to the list
for i in range(1, 97):
    dfs.append(globals()[f"df{i}"])

# concatenate all the dataframes in the list into one dataframe
all = pd.concat(dfs, ignore_index=True)
```

## Add features 2: age groups

```python
# Define the age group mapping
age_groups = {
    (25, 40): 1,
    (18, 24): 2,
    (41, 60): 3,
    (61, 99): 4
}

# Function to map age to age group
def map_age_to_group(age):
    for age_range, group in age_groups.items():
        if age_range[0] <= age <= age_range[1]:
            return group

# Apply the mapping to create the 'age_group' column
all['age_group'] = all['age'].apply(map_age_to_group)

# Display the updated DataFrame
all.head()
```

## Remove Outliers

```python
# Version WITH outliers = all2022new.csv
# We remove outliers and create: all2022new_no_outlier_csv
# Not all possible outliers are removed because it may harm the model.
# Focus is on eliminating outliers that occur in the first half of the marathon because in the second part outliers may become more 'normal'
```

```python
# Calculate the standard deviation for each specified column
std_5k_to_10k = all['perc_decay_5k_to_10k'].std()
std_10k_to_15k = all['perc_decay_10k_to_15k'].std()
std_15k_to_20k = all['perc_decay_15k_to_20k'].std()
std_20k_to_21_1k = all['perc_decay_20k_to_21_1k'].std()

# Check for outliers based on the standard deviation criterion
outliers = all[
    (all['perc_decay_5k_to_10k'] > 3 * std_5k_to_10k) |
    (all['perc_decay_5k_to_10k'] < -3 * std_5k_to_10k) |
    (all['perc_decay_10k_to_15k'] > 3 * std_10k_to_15k) |
    (all['perc_decay_10k_to_15k'] < -3 * std_10k_to_15k) |
    (all['perc_decay_15k_to_20k'] > 3 * std_15k_to_20k) |
    (all['perc_decay_15k_to_20k'] < -3 * std_15k_to_20k) |
    (all['perc_decay_20k_to_21_1k'] > 3 * std_20k_to_21_1k) |
    (all['perc_decay_20k_to_21_1k'] < -3 * std_20k_to_21_1k)
]
print(3 * std_5k_to_10k, 3 * std_10k_to_15k, 3 * std_15k_to_20k, 3 * std_20k_to_21_1k)
# Display the outliers DataFrame
print(outliers.head(20))
print(outliers.shape)

# Get the 'bib' values from outliers
outlier_bibs = outliers['bib']

# Remove rows (outliers) from 'all' based on 'bib' column
all = all[~all['bib'].isin(outlier_bibs)]

# Display the filtered DataFrame without outliers
# old shape (24485, 74) ca 1500 to be removed
print(all.shape)
all.head()
```
